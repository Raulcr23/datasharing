```python
# Importar las bibliotecas necesarias
import pandas as pd
import numpy as np

# a. Extraer la información del archivo
# Cargar el archivo CSV
df = pd.read_csv('country_vaccinations.csv')
print("Primeras filas del DataFrame:")
print(df.head())

# b. Mostrar la estructura y tipos de datos de cada columna
# Mostrar la estructura y los tipos de datos de cada columna
print("\nInformación del DataFrame:")
print(df.info())

# Convertir la columna 'date' a tipo datetime
df['date'] = pd.to_datetime(df['date'])

print("\nTipos de datos después de convertir 'date':")
print(df.dtypes)

# c. Determinar la cantidad de vacunas aplicadas de cada compañía
# Determinar la cantidad de vacunas aplicadas por cada compañía
vacunas_compania = df.groupby('vaccines')['total_vaccinations'].sum()
print("\nCantidad de vacunas aplicadas de cada compañía:")
print(vacunas_compania)

# d. Obtener la cantidad de vacunas aplicadas en todo el mundo
# Obtener la cantidad de vacunas aplicadas en todo el mundo
total_vacunas_mundo = df['total_vaccinations'].sum()
print(f"\nCantidad de vacunas aplicadas en todo el mundo: {total_vacunas_mundo} vacunas")

# e. Calcular el promedio de vacunas aplicadas por país
# Calcular el promedio de vacunas aplicadas por país
promedio_vacunas_pais = df.groupby('country')['total_vaccinations'].mean()
print("\nPromedio de vacunas aplicadas por país:")
print(promedio_vacunas_pais)

# f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo
# Filtrar la cantidad de vacunas aplicadas el día 29/01/21
vacunas_29_01_21 = df[df['date'] == '2021-01-29']['total_vaccinations'].sum()
print(f"\nCantidad de vacunas aplicadas el día 29/01/21: {vacunas_29_01_21} vacunas")

# g. Crear un dataframe nuevo denominado conDiferencias
# Crear una columna 'daily_vaccinations_raw' si no existe (se asume que es igual a 'daily_vaccinations' si no está)
df['daily_vaccinations_raw'] = df['daily_vaccinations']
# Crear el DataFrame con la diferencia de vacunas diarias
df['diferencia'] = df['daily_vaccinations'] - df['daily_vaccinations_raw']
conDiferencias = df.copy()
print("\nDataFrame con diferencias (primeras filas):")
print(conDiferencias.head())

# h. Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua
# Obtener el periodo de tiempo entre la fecha más reciente y la más antigua
fecha_max = df['date'].max()
fecha_min = df['date'].min()
diferencia_tiempo = fecha_max - fecha_min
print(f"\nPeriodo de tiempo entre la fecha más reciente y la más antigua: {diferencia_tiempo}")

# i. Crear un dataframe nuevo denominado conCantidad
# Crear una columna con la cantidad de vacunas utilizadas cada día, separando por el carácter ','
df['canVac'] = df['vaccines'].str.split(',').str.len()
conCantidad = df[['country', 'total_vaccinations', 'daily_vaccinations', 'canVac']].copy()
print("\nDataFrame con cantidad de vacunas (primeras filas):")
print(conCantidad.head())

# j. Generar un dataframe denominado antes20
# Crear un DataFrame con registros antes del 20 de diciembre de 2020
antes20 = df[df['date'] < '2020-12-20'].copy()
print("\nDataFrame con datos antes del 20 de diciembre de 2020 (primeras filas):")
print(antes20.head())

# k. Obtener un dataframe denominado pfizer
# Crear un DataFrame con registros donde se ha utilizado la vacuna Pfizer
pfizer = df[df['vaccines'].str.contains('Pfizer', na=False)].copy()
print("\nDataFrame con datos de Pfizer (primeras filas):")
print(pfizer.head())

# l. Almacenar los dataframes generados en un archivo de Excel
with pd.ExcelWriter('resultadosReto.xlsx') as writer:
    conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
    antes20.to_excel(writer, sheet_name='antes20', index=False)
    pfizer.to_excel(writer, sheet_name='pfizer', index=False)

print("\nDataFrames guardados en 'resultadosReto.xlsx'.")
```
